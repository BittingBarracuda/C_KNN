```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
```

```python
col_names = ['sample_code', 'clump_thickness', 'uniformity_cell_size', 
            'uniformity_cell_shape', 'marginal_adhesion', 'single_epithelial_cell_size',
            'bare_nuclei', 'bland_chromatin', 'normal_nucleoli', 'mitoses', 'class']
df = pd.read_csv('./breast-cancer-wisconsin.csv', names = col_names)
print(df.shape)
df.head()
```

```python
Y = df['class'].values
```

```python
df = df.drop(['sample_code', 'class'], axis=1)
df.head()
```

```python
df[df == '?'] = np.nan
df.isnull().sum()
```

```python
simple_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
col_names = col_names[1:-1]
df = pd.DataFrame(simple_imputer.fit_transform(df), columns=col_names)
df.head()
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
df = pd.DataFrame(scaler.fit_transform(df), columns=col_names)
df.head()
```

```python
X = df.values
```

```python
DATASET_FILE = 'data.txt'
CLASSES_FILE = 'classes.txt'

with open(CLASSES_FILE, mode='w') as c_f:
    rows = Y.shape[0]
    for i in range(rows):
        c_f.write(f'{Y[i]}\n')
with open(DATASET_FILE, mode='w') as d_f:
    rows, cols = X.shape
    d_f.write(f'{rows}\n')
    d_f.write(f'{cols}\n')
    for i in range(rows):
        for j in range(cols):
            d_f.write(f'{X[i][j]}\n')
```
